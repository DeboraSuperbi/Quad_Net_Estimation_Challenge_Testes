```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Lista de arquivos a serem processados
files = ['exp004a.csv', 'exp004b.csv', 'exp004c.csv']

for file_name in files:
    # Carrega os dados
    data = pd.read_csv(file_name, header=None, sep=',', dtype=float)
    
    # Detecta número de osciladores (colunas - 1 para tempo)
    num_osc = data.shape[1] - 1
    osciladores = [f'oscilador_{i+1}' for i in range(num_osc)]
    
    # Renomeia colunas
    data.columns = ['tempo'] + osciladores
    
    # ------------------ 1. Gráfico completo ------------------
    # plt.figure(figsize=(12, 5))
    # for i in range(num_osc):
    #     plt.plot(data['tempo'], data[osciladores[i]], label=f'Oscilador {i+1}', alpha=0.8)
    # plt.title(f'Sinais dos Osciladores ({file_name[:-4]}) - Completo')
    # plt.xlabel('Tempo (s)')
    # plt.ylabel('Amplitude')
    # plt.legend(ncol=2, fontsize=9)
    # plt.xlim(data['tempo'].min(), data['tempo'].max())
    # plt.grid(True, linestyle="--", alpha=0.6)
    # plt.savefig(f'{file_name[:-4]}_completo.png', dpi=300, bbox_inches='tight')
    # plt.show()

    
    # ------------------ 2. Gráfico com zoom (últimos 5s) ------------------
    plt.figure(figsize=(12, 5))
    for i in range(num_osc):
        plt.plot(data['tempo'], data[osciladores[i]], label=f'Oscilador {i+1}', alpha=0.8)
    plt.title(f'Sinais dos Osciladores ({file_name[:-4]}) - Zoom 115-120s')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude')
    plt.legend(loc='upper right', ncol=2, fontsize=9)
    
    # Define o intervalo de zoom
    tempo_max = data['tempo'].max()
    tempo_min = max(data['tempo'].min(), tempo_max - 5)  # últimos 5 segundos ou desde o início
    plt.xlim(tempo_min, tempo_max)
    
    plt.xlim(115, 120)
    plt.xticks(np.arange(115, 121, 1))
    
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.savefig(f'{file_name[:-4]}_zoom.png', dpi=300, bbox_inches='tight')
    plt.show()
```
