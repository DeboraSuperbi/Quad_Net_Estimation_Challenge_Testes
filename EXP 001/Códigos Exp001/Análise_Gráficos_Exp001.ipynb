```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Nome do arquivo a ser lido
file_name = 'exp001a.csv'

# Carrega o arquivo CSV. Como não há cabeçalho, definimos os nomes das colunas manualmente.
column_names = ['tempo', 'oscilador_1', 'oscilador_2']
data = pd.read_csv(file_name, header=None, names=column_names)

# Exibe as primeiras linhas para verificar se os dados foram carregados corretamente
print(data.head())


#########################################
# Gerar gráficos para os três arquivos
#########################################

# Lista de arquivos a serem processados
files = ['exp001a.csv', 'exp001b.csv', 'exp001c.csv']

# Loop para gerar gráficos de cada arquivo
for file_name in files:
    # Define os nomes para as colunas
    column_names = ['tempo', 'oscilador_1', 'oscilador_2']
    
    # Carrega os dados
    data = pd.read_csv(file_name, header=None, names=column_names)
    
    # Cria a figura
    plt.figure(figsize=(12, 5))
    

    # Plota os dados
    plt.plot(data['tempo'], data['oscilador_1'], label='Oscilador 1')
    plt.plot(data['tempo'], data['oscilador_2'], label='Oscilador 2', alpha=0.8)
    
    # Adiciona título e legendas
    plt.title(f'Sinais dos Osciladores ({file_name[:-4]}) - Zoom 115-120s')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude')
    plt.legend(loc="upper right")

    
    # Limites e marcações do eixo x
    plt.xlim(115, 120)
    plt.xticks(np.arange(115, 121, 1))
    
    # Grade
    plt.grid(True)
    
    # Salva a figura em arquivo PNG
    plt.savefig(f'{file_name[:-4]}.png', dpi=300)
    
    # Mostra o gráfico
    plt.show()
```
